# GRAPH --------------------------------------------------------------------------------------------------

```python
import numpy as np
import matplotlib.pyplot as plt

# Calculate the mean for each row and column
row_means = np.mean(dataset, axis=1)
col_means = np.mean(dataset, axis=0)

# Create labels for rows (e.g., 1, 2, 3, ...)
row_labels = [str(i+1) for i in range(dataset.shape[0])]

# Create labels for columns (e.g., A, B, C, ...)
col_labels = [chr(65 + i) for i in range(dataset.shape[1])]  # Assumes A=0, B=1, C=2, ...

# Generate 24 different colors for rows
row_colors = [plt.cm.viridis(i/24) for i in range(24)]
col_colors = ['lightblue', 'lightgreen', 'lightcoral', 'lightsalmon']

# Create bar graph for row means with 24 different colors
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

# Bar graph for row means
axs[0].bar(range(len(row_means)), row_means, color=row_colors)
axs[0].set_xticks(range(len(row_means)))
axs[0].set_xticklabels(row_labels, rotation=45)  # Rotate the row labels by 45 degrees
axs[0].set_xlabel('Rows')
axs[0].set_ylabel('Mean Value')
axs[0].set_title('Mean Value for Each Row')

# Annotate bars with values
for i, mean_value in enumerate(row_means):
    axs[0].text(i, mean_value, f'{mean_value:.2f}', ha='center', va='bottom', fontsize=8)

# Bar graph for column means
axs[1].bar(range(len(col_means)), col_means, tick_label=col_labels, color=col_colors)
axs[1].set_xlabel('Columns')
axs[1].set_ylabel('Mean Value')
axs[1].set_title('Mean Value for Each Column')

# Annotate bars with values
for i, mean_value in enumerate(col_means):
    axs[1].text(i, mean_value, f'{mean_value:.2f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
# Flatten the dataset to create a 1D array
flat_dataset = dataset.flatten()

# Create a histogram
plt.figure(figsize=(8, 6))
plt.hist(flat_dataset, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Dataset')
plt.grid(True)

plt.tight_layout()
plt.show()
```
